#### üìè height=4 in FacetGrid

* height=4 sets the height (in inches) of each subplot (facet) in the grid.
* The width is automatically determined by the aspect ratio (which defaults to 1), so the default subplot size is 4 inches √ó 4 inches.

* If you set: 
    * sns.FacetGrid(df, col="species", height=5)
    * Each subplot will be 5 inches tall and 5 inches wide (square), unless you change the aspect.

* You can also control the width/height ratio using aspect:
    * sns.FacetGrid(df, col="species", height=4, aspect=1.5)
    * This will make each subplot:
        * Height: 4 inches
        * Width: 4 √ó 1.5 = 6 inches

### üõ†Ô∏è Pair Plot

* A **pair plot** visualizes all **2D projections** of the dataset.
* Or we can say **Pairwise Scatter Plot** is called **Pair-Plot**
* Since we have 4 features:
    * Sepal length, Sepal width, Petal length and Petal width
    * We can create **4C2 = 6** unique 2D plots.
    * 
* Each plot shows a **pairwise** relationship between two features.
* Advantages: -
    * Can be used when number of features are high
    * Can't visulaize higher Dimensional Patters in 3D and 4D. Only possible to view 2D patterns.

### üìà How It Works

* Use `sns.pairplot(iris, hue="species")` to generate the plots.
* The result is a **matrix of plots**:
  * Rows ‚Üí y-axis labels
  * 
  * Columns ‚Üí x-axis labels
  * 
  * Diagonal ‚Üí Histograms (for now, ignore these)

* Each cell in the matrix is a **scatterplot** for a feature pair.
* You‚Äôll see **color-coded classes**:
  * Blue: Setosa
  * Orange: Versicolor
  * Green: Virginica

### üîÑ Symmetry of Pair Plots

* The plot matrix has **redundancy**:
  * Upper triangle and lower triangle are **mirror images**.
* Focus on **upper triangle (above diagonal)** to reduce redundancy.
*

## üìà 1D Scatter Plots and Beyond

### ‚ùì Introduction

* Now we ask: **What about 1D scatter plots?**

### üîπ How 1D Scatter Plot Works
* Choose a single feature (e.g. **petal length**).
* X-axis: Petal Length
* Y-axis: Set to **0 for all points** (no meaningful vertical component).

````python
# Extract subsets by species

# New DataFrame with only the subset of data for the species setosa.
iris_setosa = iris.loc[iris["species"] == "setosa"]
# New DataFrame with only the subset of data for the species versicolor.
iris_versicolor = iris.loc[iris.species == "versicolor"]
# New DataFrame with only the subset of data for the species virginica.
iris_virginica = iris[iris.species == "virginica"]

# Scatter plots with y=0
plt.plot(iris_setosa["petal length (cm)"], np.zeros_like(iris_setosa["petal length (cm)"]),'o', label='setosa')
plt.plot(iris_versicolor["petal length (cm)"], np.zeros_like(iris_versicolor["petal length (cm)"]),'o', label='versicolor')
plt.plot(iris_virginica["petal length (cm)"], np.zeros_like(iris_virginica["petal length (cm)"]),'o', label='virginica')

plt.xlabel("Petal Length (cm)")
plt.ylabel("New Zero array of Petal Length size")
plt.legend()
plt.show()
````

### ‚ö†Ô∏è Problems with 1D Scatter Plots

* **Overlapping points** make it hard to know how many data points exist at each value. You can see 50 datapoints of setosa in between 1-2 cm of PL.
* No visual count or density representation.
* Last-drawn species might hide others behind them (some orange-vercicolor are hidden behind green-verginica).


### üìä Histograms: The Fix

* Break X-axis (petal length) into **intervals or bins** (e.g., \[1.0‚Äì1.2], \[1.2‚Äì1.4], etc.)
* Y-axis: **Count of data points** in each interval
* Stack rectangles (bars) for each bin
* Such a plot is called a **Histogram**.
* In other words, Histogram is x-y plot where-
    * x-axis is corresponding to the variable you are interested in, which is Petal Length here
    * y-axis represents how often , how many times you find the data point corresponding to the given petal length.
<br>

#### 2 ways to create Histogram -

## **üü£ Cumulative Distribution Function (CDF)**

#### **‚úÖ What is CDF?**
- The Cumulative Distribution Function (CDF) of a real-valued random variable X is the function:

                            ùêπ(ùë•)=ùëÉ(ùëã‚â§ùë•)

- In simple terms, CDF gives the probability that a value from the dataset is less than or equal to x.



#### **üîµ Relation to PDF (Probability Density Function)**
- PDF tells us how densely values are packed in a small range.

- CDF tells us how many values lie below or equal to a given point.




#### **üî∏ Key Properties of CDF**
- Always starts at 0 and ends at 1

- Non-decreasing function

- Horizontal axis = feature values
Vertical axis = cumulative probability

- Useful to find the threshold values and proportion of data below a point

#### **üß† Why is CDF Useful?**
- Helps us estimate:

    - Proportion of data points below a threshold

    - Classification boundaries in simple models

- Gives more interpretable probabilities than PDFs

- Great for threshold-based decisions (e.g. "If petal length ‚â§ 2.0 ‚ûù Setosa")

#### **üîß How is CDF Computed?**
(1) Sort the data.

(2) For each point x, count how many values are ‚â§ x.

(3) Divide the count by the total number of data points.

(4) Plot x vs. this cumulative fraction.



## **üü£ CDF and PDF with the Iris Dataset**

````python
# Need for Cumulative Distribution Function (CDF)
# We can visually see what percentage of versicolor flowers have a 
# petal_length of less than 1.6?
# How to construct a CDF?
# How to read a CDF?

# Plot CDF of petal_length

counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins=10, 
                                 density=True)

pdf = counts / (sum(counts))
print(pdf)
print(bin_edges)

# compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

plt.show();
````

## **üü£ Understanding Median: A Robust Measure of Central Tendency**


**üìå Why Median?**

We‚Äôve seen that:

Mean, Variance, and Standard Deviation are sensitive to outliers.

Even a single extreme value can drastically change the mean.

So, the question is:

Is there a measure of central tendency that is resistant to outliers?

**‚úÖ Yes! That‚Äôs the Median.**

---

## **üìò Definition of Median**
The Median is the middle value in a sorted list of numbers.

- If n is odd, the median is the middle value.

- If n is even, the median is the average of the two middle values.

## **üî¢ Code Demonstration with NumPy**

````python
import numpy as np

# Example: Petal lengths of Setosa (without outlier)
setosa_petal_lengths = [1.4, 1.5, 1.3, 1.5, 1.4, 1.3, 1.5]
median_without_outlier = np.median(setosa_petal_lengths)

# Adding an extreme outlier (50)
setosa_with_outlier = setosa_petal_lengths + [50]
median_with_outlier = np.median(setosa_with_outlier)

print("Median without outlier:", median_without_outlier)
print("Median with outlier   :", median_with_outlier)
````

**‚úîÔ∏è Output: The median remains the same even after adding an outlier!**

## **üìä Comparison with Mean**

````python
mean_without_outlier = np.mean(setosa_petal_lengths)
mean_with_outlier = np.mean(setosa_with_outlier)

print("Mean without outlier:", mean_without_outlier)
print("Mean with outlier   :", mean_with_outlier)
````

**‚ùå Mean is heavily affected by the outlier.**

## **üßÆ How to Calculate Median (Step-by-Step)**

### **‚ñ∂Ô∏è Case 1: Odd Number of Elements**

````python
# Given data (unsorted)
x = [1, 1.2, 1.1, 2.1, 1.8, 1.6, 1.2]

# Step 1: Sort the data
x_sorted = sorted(x)

# Step 2: Find the middle value
n = len(x_sorted)
median = x_sorted[(n // 2)]  # 0-based indexing

print("Sorted data:", x_sorted)
print("Median (odd n):", median)
````

### **‚ñ∂Ô∏è Case 2: Even Number of Elements**

````python
# Adding one more value to make it even
x_even = x + [1.8]
x_even_sorted = sorted(x_even)
n_even = len(x_even_sorted)

# Median = average of n/2 and (n/2 - 1) indices
median_even = (x_even_sorted[n_even // 2 - 1] + x_even_sorted[n_even // 2]) / 2

print("Sorted data (even):", x_even_sorted)
print("Median (even n):", median_even)
````

## **‚ö†Ô∏è Why is Median Robust to Outliers?**

````python
# Original data
x_clean = [1, 1.1, 1.2, 1.4, 1.6, 1.8, 2.0]
median_clean = np.median(x_clean)

# Add one outlier
x_with_outlier = x_clean + [56]
median_with_outlier = np.median(x_with_outlier)

print("Median without outlier:", median_clean)
print("Median with one outlier:", median_with_outlier)
````

**‚úÖ Result: Median barely changes**

**‚ùå Mean would be highly affected**

### **‚ö†Ô∏è When Median Fails? (Majority Corrupted)**

````python
# Majority corrupted data
x_bad = [1, 1.1, 1.2, 10, 20, 30, 40]
median_bad = np.median(x_bad)

print("Corrupted data:", x_bad)
print("Median (with majority corrupted):", median_bad)
````

### **‚úÖ Summary: Median vs Mean**


| Feature              | Mean  | Median          |
| -------------------- | ----- | --------------- |
| Affected by Outliers | ‚úÖ Yes | ‚ùå No (if < 50%) |
| Easy to Compute      | ‚úÖ Yes | ‚úÖ Yes           |
| Represents Center    | ‚úÖ Yes | ‚úÖ Yes           |
| Robustness           | ‚ùå Low | ‚úÖ High          |


### **üìå Conclusion**

- Median is a robust alternative to the mean.

- It‚Äôs simple to calculate: just sort the data and pick the middle value.

- Ideal for datasets with outliers or skewed distributions.

### **üìå 3. Median Absolute Deviation (MAD)**

MAD measures the typical distance of data points from the median, not the mean.
It‚Äôs robust and not affected by extreme values (outliers).

**Steps to calculate MAD:**

1. Find the median of the data.

2. Compute absolute deviation from median: |xi - median|

3. Find the median of all absolute deviations.

**Formula:**
MAD=median(‚à£xi‚àímedian(x)‚à£)

- 25% of values < Q1

### **üìå 6. Why Use MAD and IQR?**

# üì¶ Understanding Box Plots and Whiskers in Data Visualization

- From CDFs, we can identify percentile values by drawing horizontal and vertical lines.

### ‚ùì Can we get percentile values more easily?
Yes! ‚û§ **Box plots** are designed to represent percentiles directly and clearly.

A **box plot** (also known as a **box-and-whisker plot**) uses percentiles to summarize the distribution of a dataset.

# üìä Violin Plot: Merging Box Plot and PDF

In data visualization, we‚Äôve already explored:
- **Histograms** and **Probability Density Functions (PDFs)**
- **Box Plots**

Each has its own strengths:
- üì¶ **Box Plot** helps visualize percentiles and outliers.
- üìà **PDF/Histogram** shows the distribution shape.

But what if we want the **best of both worlds**?  
üëâ Enter the **Violin Plot**.

---

## üéª What is a Violin Plot?

A **violin plot** combines the information of:
- **Box plots** (percentiles, whiskers, median)
- **Density plots** (smooth histogram/PDF)

The name comes from its **violin-like shape** üéª.

---

## üîç Components of a Violin Plot

Here's what you see in a violin plot:

### 1. **Box Plot (Inside the Violin)**
- The **thick black box** in the center is the traditional box plot.
- Contains:
  - **White dot**: Median (50th percentile)
  - **Top edge of box**: 75th percentile (Q3)
  - **Bottom edge of box**: 25th percentile (Q1)
  - **Whiskers**: Data range (typically Q1 - 1.5*IQR to Q3 + 1.5*IQR)

### 2. **Density Plot (Outer Shape)**
- The **violin shape** on either side of the box is the **PDF** (Probability Density Function).
- Symmetric and smooth ‚Äî like a **bell curve** (if Gaussian).

This gives us insight into both:
- **Data distribution** (shape and spread)
- **Statistical summary** (via box plot)

---

## **üîπ 3. Types of Data Analysis**

### **üìä Univariate Analysis
- Analyzing one variable at a time

- Examples:

    - Histograms

    - Box plots

    - Violin plots

    - PDFs, CDFs

Useful for understanding distribution, central tendency, spread, and outliers.

### **üìâ Bivariate Analysis**
- Analyzing two variables together

- Examples:

    - Scatter plots

    - Pair plots (2D combinations)

Helps identify relationships or correlations between two features (e.g., petal length vs petal width).


### **üìà Multivariate Analysis**
- Analyzing three or more variables

- Example:

    - 3D scatter plots

    - Any ML model using multiple features

All of machine learning can be viewed as an advanced form of multivariate analysis.
